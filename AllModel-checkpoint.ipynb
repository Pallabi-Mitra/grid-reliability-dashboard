```python
# Cell 1: install dependencies
import subprocess
subprocess.run(["pip", "install", "xgboost==2.1.4", "--quiet"])
subprocess.run(["pip", "install", "lightgbm", "joblib", "--quiet"])
```

```python
#Cell 2: imports and data loading
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score

assets = pd.read_csv("assets.csv")
daily = pd.read_csv("daily_records.csv")
df = daily.merge(assets, on="asset_id", how="left")

print(f"XGBoost version: {__import__('xgboost').__version__}")
print(f"Total rows: {len(df)}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"Generators: {df['asset_id'].nunique()}")
print(f"Zones: {df['operating_region'].nunique()}")
```

```python
#Cell 3: feature engineering
categorical_cols = ["season", "fuel_category", "broad_asset_category", "operating_region"]

df_encoded = pd.get_dummies(df, columns=categorical_cols)
df_encoded["date"] = df["date"].values

y = df_encoded["impact_ratio"]

print(f"Target range: {y.min():.3f} to {y.max():.3f}, mean: {y.mean():.3f}")
print(f"Encoded shape: {df_encoded.shape}")
```

```python
#Cell 4: time-based train/test split
df_encoded_sorted = df_encoded.sort_values("date").reset_index(drop=True)

split_idx = int(len(df_encoded_sorted) * 0.88)

train = df_encoded_sorted.iloc[:split_idx]
test = df_encoded_sorted.iloc[split_idx:]

# Features: exclude non-model columns
exclude_cols = ["date", "asset_id", "impact_ratio", "date_str"]
feature_cols = [c for c in df_encoded_sorted.columns if c not in exclude_cols
                and df_encoded_sorted[c].dtype in [np.float64, np.int64, np.float32, np.int32, bool]]

X_train = train[feature_cols]
y_train = train["impact_ratio"]
X_test = test[feature_cols]
y_test = test["impact_ratio"]

print(f"Train rows: {len(X_train)}, Test rows: {len(X_test)}")
print(f"Features: {len(feature_cols)}")
print(f"Train date range: {train['date'].iloc[0]} to {train['date'].iloc[-1]}")
print(f"Test date range: {test['date'].iloc[0]} to {test['date'].iloc[-1]}")
```

```python
#Cell 5: train main XGBoost model
model = XGBRegressor(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    objective="reg:squarederror"
)

model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=50)

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nMain model results:")
print(f"  MAE:  {mae:.4f}")
print(f"  R²:   {r2:.4f}")
```

```python
#Cell 6: verify SHAP works before saving
import shap

explainer = shap.TreeExplainer(model)
sample = X_test.iloc[:5]
shap_vals = explainer.shap_values(sample)

print("SHAP test passed, shape:", shap_vals.shape)
print("Top 5 features by global importance:")
importance = pd.Series(
    np.abs(shap_vals).mean(axis=0),
    index=feature_cols
).sort_values(ascending=False).head(5)
print(importance)
```

```python
#Cell 7: save main model
model.save_model("model.json")
print("Saved: model.json")
print(f"Feature names: {model.get_booster().feature_names[:5]}...")
```

```python
#Cell 8: train three quantile forecaster models
quantiles = {
    "p05": 0.05,
    "p50": 0.50,
    "p95": 0.95,
}

forecaster_models = {}

for name, alpha in quantiles.items():
    print(f"Training forecaster_{name} (alpha={alpha})...")
    m = XGBRegressor(
        objective="reg:quantileerror",
        quantile_alpha=alpha,
        n_estimators=300,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    m.fit(X_train, y_train)
    forecaster_models[name] = m

    test_mae = mean_absolute_error(y_test, m.predict(X_test))
    print(f"  Test MAE: {test_mae:.4f}")

print("\nAll forecaster models trained.")
```

```python
#Cell 9: save forecaster models
for name, m in forecaster_models.items():
    filename = f"forecaster_{name}.json"
    m.save_model(filename)
    print(f"Saved: {filename}")
```

```python
#Cell 10: Train LightGBM
import lightgbm as lgb

lgb_model = lgb.LGBMRegressor(
    n_estimators=300, learning_rate=0.05,
    num_leaves=31, random_state=42, verbose=-1
)
lgb_model.fit(X_train, y_train)
lgb_preds = lgb_model.predict(X_test)
print(f"LightGBM MAE: {mean_absolute_error(y_test, lgb_preds):.4f}")
print(f"LightGBM R²: {r2_score(y_test, lgb_preds):.4f}")
lgb_model.booster_.save_model("model_lightgbm.txt")
print("Saved: model_lightgbm.txt")
```

```python
#Cell 11: Train Random Forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=200, max_depth=10,
    random_state=42, n_jobs=-1
)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)
print(f"Random Forest MAE: {mean_absolute_error(y_test, rf_preds):.4f}")
print(f"Random Forest R²: {r2_score(y_test, rf_preds):.4f}")

import joblib
joblib.dump(rf_model, "model_random_forest.pkl")
print("Saved: model_random_forest.pkl")
```

```python
#Cell 12: Train Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor

gbm_model = GradientBoostingRegressor(
    n_estimators=200, max_depth=4,
    learning_rate=0.05, random_state=42
)
gbm_model.fit(X_train, y_train)
gbm_preds = gbm_model.predict(X_test)
print(f"Gradient Boosting MAE: {mean_absolute_error(y_test, gbm_preds):.4f}")
print(f"Gradient Boosting R²: {r2_score(y_test, gbm_preds):.4f}")

joblib.dump(gbm_model, "model_gradient_boosting.pkl")
print("Saved: model_gradient_boosting.pkl")
```

```python
#Cell 10: sanity check all 4 models
# Reload and verify all models work
main_check = XGBRegressor()
main_check.load_model("model.json")

f05 = XGBRegressor(); f05.load_model("forecaster_p05.json")
f50 = XGBRegressor(); f50.load_model("forecaster_p50.json")
f95 = XGBRegressor(); f95.load_model("forecaster_p95.json")

sample_row = X_test.iloc[[0]]

main_pred = main_check.predict(sample_row)[0]
p05_pred = f05.predict(sample_row)[0]
p50_pred = f50.predict(sample_row)[0]
p95_pred = f95.predict(sample_row)[0]
actual = y_test.iloc[0]

print("All models reload correctly.")
print(f"\nSample prediction:")
print(f"  Main model:      {main_pred:.3f}")
print(f"  Forecaster p05:  {p05_pred:.3f}")
print(f"  Forecaster p50:  {p50_pred:.3f}")
print(f"  Forecaster p95:  {p95_pred:.3f}")
print(f"  Actual value:    {actual:.3f}")
print(f"\nSanity check p05 <= p50 <= p95: {p05_pred <= p50_pred <= p95_pred}")

# Verify SHAP still works with reloaded model
explainer2 = shap.TreeExplainer(main_check)
shap_check = explainer2.shap_values(sample_row)
print(f"SHAP on reloaded model: passed, shape {shap_check.shape}")
```
