```python
#Cell 1: install the right version first
import subprocess
subprocess.run(["pip", "install", "xgboost==2.1.4", "--quiet"])
```

```python
#Cell 2: imports and data loading
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score

assets = pd.read_csv("assets.csv")
daily = pd.read_csv("daily_records.csv")
df = daily.merge(assets, on="asset_id", how="left")

print(f"XGBoost version: {__import__('xgboost').__version__}")
print(f"Total rows: {len(df)}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"Generators: {df['asset_id'].nunique()}")
print(f"Zones: {df['operating_region'].nunique()}")
```

```python
#Cell 3: feature engineering
categorical_cols = ["season", "fuel_category", "broad_asset_category", "operating_region"]

df_encoded = pd.get_dummies(df, columns=categorical_cols)
df_encoded["date"] = df["date"].values

y = df_encoded["impact_ratio"]

print(f"Target range: {y.min():.3f} to {y.max():.3f}, mean: {y.mean():.3f}")
print(f"Encoded shape: {df_encoded.shape}")
```

```python
#Cell 4: time-based train/test split
df_encoded_sorted = df_encoded.sort_values("date").reset_index(drop=True)

split_idx = int(len(df_encoded_sorted) * 0.88)

train = df_encoded_sorted.iloc[:split_idx]
test = df_encoded_sorted.iloc[split_idx:]

# Features: exclude non-model columns
exclude_cols = ["date", "asset_id", "impact_ratio", "date_str"]
feature_cols = [c for c in df_encoded_sorted.columns if c not in exclude_cols
                and df_encoded_sorted[c].dtype in [np.float64, np.int64, np.float32, np.int32, bool]]

X_train = train[feature_cols]
y_train = train["impact_ratio"]
X_test = test[feature_cols]
y_test = test["impact_ratio"]

print(f"Train rows: {len(X_train)}, Test rows: {len(X_test)}")
print(f"Features: {len(feature_cols)}")
print(f"Train date range: {train['date'].iloc[0]} to {train['date'].iloc[-1]}")
print(f"Test date range: {test['date'].iloc[0]} to {test['date'].iloc[-1]}")
```

```python
#Cell 5: train main XGBoost model
model = XGBRegressor(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    objective="reg:squarederror"
)

model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=50)

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nMain model results:")
print(f"  MAE:  {mae:.4f}")
print(f"  R²:   {r2:.4f}")
```

```python
#Cell 6: verify SHAP works before saving
import shap

explainer = shap.TreeExplainer(model)
sample = X_test.iloc[:5]
shap_vals = explainer.shap_values(sample)

print("SHAP test passed, shape:", shap_vals.shape)
print("Top 5 features by global importance:")
importance = pd.Series(
    np.abs(shap_vals).mean(axis=0),
    index=feature_cols
).sort_values(ascending=False).head(5)
print(importance)
```

```python
#Cell 7: save main model
model.save_model("model.json")
print("Saved: model.json")
print(f"Feature names: {model.get_booster().feature_names[:5]}...")
```

```python
#Cell 8: train three quantile forecaster models
quantiles = {
    "p05": 0.05,
    "p50": 0.50,
    "p95": 0.95,
}

forecaster_models = {}

for name, alpha in quantiles.items():
    print(f"Training forecaster_{name} (alpha={alpha})...")
    m = XGBRegressor(
        objective="reg:quantileerror",
        quantile_alpha=alpha,
        n_estimators=300,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    m.fit(X_train, y_train)
    forecaster_models[name] = m

    test_mae = mean_absolute_error(y_test, m.predict(X_test))
    print(f"  Test MAE: {test_mae:.4f}")

print("\nAll forecaster models trained.")
```

```python
#Cell 9: save forecaster models
for name, m in forecaster_models.items():
    filename = f"forecaster_{name}.json"
    m.save_model(filename)
    print(f"Saved: {filename}")
```
