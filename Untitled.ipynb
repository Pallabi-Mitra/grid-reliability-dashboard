```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
import json

# Load the same data your main model uses
assets = pd.read_csv("assets.csv")
daily = pd.read_csv("daily_records.csv")
df = daily.merge(assets, on="asset_id", how="left")

print(f"Total rows: {len(df)}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
```

```python
categorical_cols = ["season", "fuel_category", "broad_asset_category", "operating_region"]

df_encoded = pd.get_dummies(df, columns=categorical_cols)

# Target
y = df_encoded["impact_ratio"]

# Features: same as main model
main_model = XGBRegressor()
main_model.load_model("model.json")
model_features = main_model.get_booster().feature_names

# Align columns
for col in model_features:
    if col not in df_encoded.columns:
        df_encoded[col] = 0

X = df_encoded[model_features]

print(f"Features: {len(model_features)}")
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

```python
# Sort by date, use last 12% of rows as test (same ~90 day split as main model)
df_encoded["date"] = df["date"].values
df_encoded_sorted = df_encoded.sort_values("date").reset_index(drop=True)

split_idx = int(len(df_encoded_sorted) * 0.88)

X_train = df_encoded_sorted[model_features].iloc[:split_idx]
y_train = df_encoded_sorted["impact_ratio"].iloc[:split_idx]
X_test = df_encoded_sorted[model_features].iloc[split_idx:]
y_test = df_encoded_sorted["impact_ratio"].iloc[split_idx:]

print(f"Train rows: {len(X_train)}, Test rows: {len(X_test)}")
print(f"Train date range: {df_encoded_sorted['date'].iloc[0]} to {df_encoded_sorted['date'].iloc[split_idx-1]}")
print(f"Test date range: {df_encoded_sorted['date'].iloc[split_idx]} to {df_encoded_sorted['date'].iloc[-1]}")
```

```python
quantiles = {
    "p05": 0.05,   # low/optimistic estimate
    "p50": 0.50,   # median/typical estimate
    "p95": 0.95,   # high/pessimistic estimate
}

models = {}

for name, alpha in quantiles.items():
    print(f"Training {name} (alpha={alpha})...")
    model = XGBRegressor(
        objective="reg:quantileerror",
        quantile_alpha=alpha,
        n_estimators=300,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    model.fit(X_train, y_train)
    models[name] = model
    print(f"  Done. Test MAE: {abs(model.predict(X_test) - y_test).mean():.4f}")

print("\nAll 3 quantile models trained.")
```

```python
for name, model in models.items():
    filename = f"forecaster_{name}.json"
    model.save_model(filename)
    print(f"Saved: {filename}")

print("\nAll forecaster models saved to project root.")
```

```python
# Verify the models load and produce sensible output
test_row = X_test.iloc[[0]]

p05 = models["p05"].predict(test_row)[0]
p50 = models["p50"].predict(test_row)[0]
p95 = models["p95"].predict(test_row)[0]

print(f"Sample prediction for one generator:")
print(f"  Low estimate (p05):    {p05:.3f}")
print(f"  Typical estimate (p50): {p50:.3f}")
print(f"  High estimate (p95):   {p95:.3f}")
print(f"  Actual value:          {y_test.iloc[0]:.3f}")
print(f"\nBand width (p95-p05): {p95-p05:.3f}")
print("Sanity check: p05 <= p50 <= p95?", p05 <= p50 <= p95)
```
